```python
##Answer to question 2.1.1a

from scipy import stats
import scipy.io
import matplotlib.pyplot as plt
import numpy as np

wine_mat = scipy.io.loadmat('./Data/wine.mat')
data = wine_mat['X']
s_data = stats.zscore(data)
attribute_names = wine_mat['attributeNames']

# print(len(wine_mat['attributeNames'][0]))
# for i in wine_mat['attributeNames'][0]:
#     print(i[0])
# for i in data[0]:
#     print('{0}'.format(i))

fig1, axes1 = plt.subplots(4,3, figsize=(10,20))
fig2, axes2 = plt.subplots(4,3, figsize=(10,20))

for i in range(4):
    for j in range(3):
        axes1[i,j].boxplot(s_data[:,(i*3)+j])
        axes1[i,j].set_title(attribute_names[0][(i*3)+j][0])
        
for i in range(4):
    for j in range(3):
        axes2[i,j].hist(data[:,(i*3)+j])
        axes2[i,j].set_title(attribute_names[0][(i*3)+j][0])

volatile_acidity = data[:,1]
va_mask = volatile_acidity<20
clean_volatile_acidity = volatile_acidity[va_mask]

density = data[:,7]
d_mask = density<10
clean_density = density[d_mask]

alcohol = data[:,10]
a_mask = alcohol<200
clean_a = alcohol[a_mask]

fig3, axes3 = plt.subplots(nrows=1,ncols=3, figsize=(10,10))
axes3[0].boxplot(clean_volatile_acidity)
axes3[0].set_title('Volatile Acidity')
axes3[1].boxplot(clean_density)
axes3[1].set_title('Density')
axes3[2].boxplot(clean_a)
axes3[2].set_title('Alcohol')


```

> 2.1.2 Make scatter plots between attributes and wine quality as rated by human judges. Can you manually identify any clear relationship between the attributes of the wine and wine quality? Which values of these attributes are associated with high quality wine? Use the correlation coefficients to substantiate your answers.
+ *You can calculate the correlation coefficient using the scipy.stats.pearsonr() function to measure the strength of association.*

```python
##Answer to question 2.1.2
from scipy import stats
import scipy.io
import matplotlib.pyplot as plt
import numpy as np

wine_mat = scipy.io.loadmat('./Data/wine.mat')
data = wine_mat['X']
s_data = stats.zscore(data)
attribute_names = wine_mat['attributeNames']

fig4, axes4 = plt.subplots(4,3, figsize=(30,20))

        
for i in range(4):
    for j in range(3):
        axes4[i,j].scatter(s_data[:,(i*3)+j],s_data[:,11])
        axes4[i,j].set_title(attribute_names[0][(i*3)+j][0])
        print("pearon's r for quality and ", attribute_names[0][(i*3)+j][0], " is ", stats.pearsonr(s_data[:,(i*3)+j],s_data[:,11]))
        
volatile_acidity = data[:,1]
va_mask = volatile_acidity<20
clean_volatile_acidity = volatile_acidity[va_mask]

density = data[:,7]
d_mask = density<10
clean_density = density[d_mask]

alcohol = data[:,10]
a_mask = alcohol<200
clean_a = alcohol[a_mask]

fig5, axes5 = plt.subplots(nrows=3,ncols=1, figsize=(20,10))
axes5[0].scatter(clean_volatile_acidity,data[:,11][va_mask])
axes5[0].set_title('Volatile Acidity')
axes5[1].scatter(clean_density, data[:,11][d_mask])
axes5[1].set_title('Density')
axes5[2].scatter(clean_a, data[:,11][a_mask])
axes5[2].set_title('Alcohol')

print("\nFor data with outliers removed:")
print("Pearon's r for quality and Volatile Acidity is ", stats.pearsonr(clean_volatile_acidity,data[:,11][va_mask]))
print("Pearon's r for quality and Density is ", stats.pearsonr(clean_density, data[:,11][d_mask]))
print("Pearon's r for quality and Alcohol is ", stats.pearsonr(clean_a, data[:,11][a_mask]))
```

In the raw data, chlorides has the highest Pearson's r. It is negative, therefore less chloride is associated with higher quality. 

In the data with outliers removed, we can see that Volatile Acidity has some negative correlation with quality. Furthermore, after inspecting Pearson's r, we can see that all three of alcohol, volatile acidity, and density correlate at least weakly with quality.
